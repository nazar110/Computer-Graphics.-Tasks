## Плетене мереживо 

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import matplotlib
import math
```

```python
R = 10
print(f"R = {R}")
```

```python
N = 10
print(f"N = {N}")
```

```python
def fibonacci(elem):
    prev = 1
    curr = 1
    res_list = [0, 1, 1]
    for i in range(1, elem - 2):
        next_ = prev + curr
        prev = curr
        curr = next_
        res_list.append(curr)
    return res_list
```

```python
fib_numbers = fibonacci(30)
for i in fib_numbers:
    print(i, end=" ")
```

```python
def get_all_points():
    all_points = [[0, R]]
    for i in range(1, N):
        all_points.append([i*R/N, R])
    all_points.append([R, R])
    temp_list = [i for i in range(1, N)]
    temp_list.reverse()
    for i in temp_list:#range(1, N):
        all_points.append([R, i*R/N])
    all_points.append([R, 0])
    for i in temp_list:
        all_points.append([i*R/N, 0])
    all_points.append([0, 0])
    for i in range(1, N):
        all_points.append([0, i*R/N])
    return all_points
# зробив послідовно, щоб можна було нумерувати точки за годинниковою стрілкою
```

```python
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(7, 7)

all_points = get_all_points()

for i in all_points:
    plt.plot(i[0], i[1], 'b.')
```

```python
for i in range(0, 4*N):
    print(all_points[i], end= " ")
```

```python
def plot_lace(N):
    for i in range(0, 4*N):
        for j in range(0, 4*N):
            sample = ((j - i)) % (4*N)
            for num in fib_numbers:
                if sample == num and sample < 4 * N:
                    plt.plot([all_points[i][0], all_points[j][0]], [all_points[i][1], all_points[j][1]])
                    # [x1, x2], [y1, y2]
```

```python
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(8, 8)
plot_lace(N)
```

```python
def see_algorithm(N):
    for i in range(0, 4*N):
        for j in range(0, 4*N):
            sample = ((j - i)) % (4*N)
            for num in fib_numbers:
                if i == 32 and num == sample  and sample < N*4:
                    plt.plot([all_points[i][0], all_points[j][0]], [all_points[i][1], all_points[j][1]])
                    print(f'{j}', end=" ")
```
